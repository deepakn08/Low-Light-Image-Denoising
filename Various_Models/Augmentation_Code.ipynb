```python
import os
import random
import numpy as np
import pandas as pd
import cv2
from PIL import Image
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import tensorflow as tf
```

```python
base_dir='/content/drive/MyDrive/Train/Train'
high_path=os.path.join(base_dir,'high')
low_path=os.path.join(base_dir,'low')
```

```python
project_path=os.makedirs(os.path.join(base_dir,'Image Denoising Project'))
```

```python
image_list_high=os.listdir(high_path)
image_list_low=os.listdir(low_path)
```

```python
len(image_list_low)
```

```python
class Image_Augmentation:
  def __init__(self):
    self.name=None
    self.high_path=None
    self.low_path=None
    self.image_high=None
    self.image_low=None

  def read_image(self,path,name):
    self.name=name
    self.high_path=os.path.join(path,'high')
    self.low_path=os.path.join(path,'low')
    self.image_high=Image.open(os.path.join(self.high_path,self.name))
    self.image_low=Image.open(os.path.join(self.low_path,name))

  def _up_down_flip(self):
    high = Image.fromarray(np.array(tf.image.flip_up_down(self.image_high)))
    low = Image.fromarray(np.array(tf.image.flip_up_down(self.image_low)))
    new_name=self.name[:-4]+'ud.png'
    high.save(os.path.join(self.high_path,new_name))
    low.save(os.path.join(self.low_path,new_name))

  def _left_right_flip(self):
    high = Image.fromarray(np.array(tf.image.flip_left_right(self.image_high)))
    low = Image.fromarray(np.array(tf.image.flip_left_right(self.image_low)))
    new_name=self.name[:-4]+'lr.png'
    high.save(os.path.join(self.high_path,new_name))
    low.save(os.path.join(self.low_path,new_name))

  def _rotate(self):
    random_angle = tf.random.uniform(shape=[], minval=0, maxval=4, dtype=tf.int32)
    high = Image.fromarray(np.array(tf.image.rot90(self.image_high,random_angle)))
    low = Image.fromarray(np.array(tf.image.rot90(self.image_low,random_angle)))
    new_name=self.name[:-4]+'rot.png'
    high.save(os.path.join(self.high_path,new_name))
    low.save(os.path.join(self.low_path,new_name))
```

```python
augmenter=Image_Augmentation()
for i,name in enumerate(image_list_high):
  augmenter.read_image(base_dir,name)
  augmenter._up_down_flip()
  augmenter._left_right_flip()
  augmenter._rotate()
  print(i)
```
